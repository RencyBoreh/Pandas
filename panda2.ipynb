```python
import pandas as pd

data = {
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C', 'A'],
    'Quantity': [5, 10, 2, 7, 8, 5, 3, 4, 6, 9],
    'TotalPrice': [50, 100, 20, 70, 80, 50, 30, 40, 60, 90]
}
df = pd.DataFrame(data)
df.to_csv('sales_data.csv', index=False)
```

```python
sorted_df = df.sort_values(by='TotalPrice', ascending=False)
sorted_df
```

```python
total_quantity_by_product = df.groupby('Product')['Quantity'].sum()
print(total_quantity_by_product)
```

```python
average_price_by_product = df.groupby('Product')['TotalPrice'].mean()
print(average_price_by_product)
```

```python
import matplotlib.pyplot as plt

total_sales_by_product = df.groupby('Product')['TotalPrice'].sum()
total_sales_by_product.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.title('Total Sales by Product')
plt.show()
```

```python
import seaborn as sns

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
